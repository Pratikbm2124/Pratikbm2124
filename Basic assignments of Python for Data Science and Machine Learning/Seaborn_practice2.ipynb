```python
import seaborn as sn
import matplotlib.pyplot as plt
iris = sn.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sn.pairplot(iris)
```

# Grid

```python
g = sn.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sn.PairGrid(iris)
g.map_upper(plt.scatter)
g.map_diag(sn.distplot)
g.map_lower(sn.kdeplot)
```

```python
tips = sn.load_dataset('tips')
tips.head()
```

```python
g1 = sn.FacetGrid(data = tips, col = 'time', row = 'smoker')
```

```python
tips['time'].unique()
```

```python
g1 = sn.FacetGrid(data = tips, col = 'time', row = 'smoker')
g1.map(sn.distplot, 'total_bill')
```

```python
g2 = sn.FacetGrid(data = tips, col = 'time', row = 'smoker')
g2.map(plt.scatter, 'total_bill', 'tip')
```

# Regression Plots

```python
tips.head()
```

```python
sn.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers=['o','v'])
```

```python
sn.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'sex', row = 'time')
```

```python
sn.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'day', hue = 'sex')
```

```python
sn.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'day', hue = 'sex',
         aspect = 0.5, height = 8)
```
