# Categorical Plots

```python
import numpy as np
```

```python
sn.barplot(x='sex', y='total_bill', data= tips)
```

```python
sn.barplot(x='sex', y='total_bill', data= tips, estimator=np.std)
```

```python
sn.countplot(x = 'sex', data = tips)
```

```python
sn.boxplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sn.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
```

```python
sn.violinplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sn.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
sn.stripplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sn.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = False)
```

```python
sn.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', dodge = True) # split renamed to dodge
# not to use for large datasets
```

```python
sn.swarmplot(x = 'day', y = 'total_bill', data = tips) 
```

```python
sn.violinplot(x = 'day', y = 'total_bill', data = tips)
sn.swarmplot(x = 'day', y = 'total_bill', data = tips, color = 'black') 
```

```python
sn.factorplot(x = 'day', y = 'total_bill', data = tips, kind = 'bar') # `factorplot` function has been renamed to `catplot`
```

```python
sn.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'bar') # `factorplot` function has been renamed to `catplot`
```

# Matrix Plots

```python
import seaborn as sn
flights = sn.load_dataset('flights')
flights.head()
```

```python
tc = tips.corr()
```

```python
sn.heatmap(tc, annot = True, cmap = 'coolwarm')
```

```python
tc
```

```python
flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
fp = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
sn.heatmap(fp)
```

```python
sn.heatmap(fp, linecolor = 'white', linewidths=1)
```

```python
sn.heatmap(fp,cmap = 'coolwarm', linecolor = 'black', linewidths=1)
```

```python
sn.clustermap(fp)
```

```python
sn.clustermap(fp, cmap = 'coolwarm')
```

```python
sn.clustermap(fp, cmap = 'coolwarm', standard_scale=1)
```
