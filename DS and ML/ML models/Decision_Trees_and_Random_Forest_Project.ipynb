```python
dummy_val = pd.get_dummies(df['purpose'], drop_first=True, prefix='purpose')
```

```python
dummy_val
```

```python
final_df = pd.concat([df, dummy_val], axis = 1)
```

```python
final_df.drop('purpose', axis = 1, inplace = True)
```

```python
final_df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df.drop('not.fully.paid', axis = 1)
y = final_df['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
pred = dt.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=300)
```

```python
rfc.fit(X_train, y_train)
```

```python
pred2 = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test, pred2))
print('\n')
print(classification_report(y_test, pred2))
```
